```python
import pandas as pd

df = pd.read_csv("/Users/rahuldandona/Desktop/KJSCE/SEM3/HO-FDS/Exp1/weatherAUS - weatherAUS.csv")
df = df.dropna()
df = df.sort_values(by=['MaxTemp'])
```

```python
df.head()
```

```python
MAX_TEMP = df["MaxTemp"].tolist()
```

## MEAN

```python
sum = 0
for i in MAX_TEMP:
    sum = sum + i
mean = sum/len(MAX_TEMP)
mean
```

## MEDIAN

```python
MAX_TEMP
```

```python
if (len(MAX_TEMP)%2==0):
    median = (MAX_TEMP[len(MAX_TEMP)//2])+ (MAX_TEMP[len(MAX_TEMP)//2-1])/2          
else:
    median = MAX_TEMP[len(MAX_TEMP)]//2
median
```

## MODE

```python
char_freq={}

for i in MAX_TEMP[:len(MAX_TEMP)+1]:
    if i in char_freq:
        char_freq[i]=char_freq[i]+1
    else:
        char_freq[i] = 1
result= max(char_freq, key = char_freq.get)
result
```

# Variance

```python
len(MAX_TEMP)
```

```python
var = 0
for i in range(len(MAX_TEMP)):
    
    var += ((MAX_TEMP[i]-mean)**2)
var = var/len(MAX_TEMP)
var 
```

# Standard Deviation

```python
sd = var**0.5
sd
```

# Mean Deviation

```python
md = 0
for i in range(len(MAX_TEMP)):
    if mean< MAX_TEMP[i]:
        md +=MAX_TEMP[i]-mean
    else:
        md += mean-MAX_TEMP[i]
     
md = md/len(MAX_TEMP)
md 
```

# Coefficient of Variation

```python
cov=(sd)/(mean)
cov
```

# Standard Error

```python
SE= (sd)/((len(MAX_TEMP))**(0.5))
SE
```
